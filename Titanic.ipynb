```python
import pandas as pd 
```

```python
df = pd.read_csv('titanic.csv', index_col = False )
```

```python
df.columns
```

```python
df_n = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis= 'columns')
```

```python
df_n
```

```python
# Converting Sex column into numerical values 
# LabelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
input = df_n.drop(['Survived'], axis = 'columns')
target = df_n['Survived']
```

```python
input
```

```python
target
```

```python
le_sex = LabelEncoder()
```

```python
input['sex_column_1'] = le_sex.fit_transform(input['Sex'])
```

```python
input
```

```python
input_n = input.drop(['Sex'], axis= 'columns')
```

```python
input_n
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split (input_n, target, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn import tree
```

```python
reg = tree.DecisionTreeClassifier()
```

```python
reg.fit(x_train, y_train)
```

```python
y_predict = reg.predict(x_test)
```

```python
reg.score(x_test, y_test)
```
